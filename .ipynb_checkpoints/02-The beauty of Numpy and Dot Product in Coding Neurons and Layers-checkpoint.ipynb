<h3 style="background-color:purple; color:white; padding:10px; border-radius:8px; text-align:center;">
  Single Neuron with Numpy
</h3>


```python
#!pip install numpy
```

```python
import numpy as np
```

```python
#Manually Garda

#Inputs , Bias ko List Banaune

#Neuron with 3 inputs
inputs = [1,2,3]
#Here x1 is 1 , x2 is 2 & x3 is 3
weights = [0.6 , 0.7 , 0.2 ] #Here w1 is 0.6 , w2 is 0.7 & w3 is 0.2
bias = 2

output = (inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + bias)
#List ko Indexing starts from 0 so x1 = inputs[0]
print(output)
```

```python
#Numpy ko dot() function use garda

inputs = [1,2,3]
weights = [0.6 , 0.7 , 0.2 ]
bias = 2

output = np.dot(inputs , weights) + bias
print(output)
```

<h3 style="background-color:purple; color:white; padding:10px; border-radius:8px; text-align:center;">
     Layer of Neurons using Numpy
</h3>


```python
inputs = [ 4 , 5 , 6 , 7]
weights = [
    #Every Neuron Sanga ko Afnai Associated Weights cha
    [ 0.3 , 0.4 , 0.9 , 0.1 ] , #Neuron1 sanga associated weights
    [ 0.4 , 0.5 , 0.1, 0.8 ],
    [0.5, 0.7 , 0.8, 0.4  ],
]
biases = [ 2, 6 , 1]

"""
inputs is Our Vector
weights is Our Matrix

so Each Weight ko Row le Each Neuron sanga associated Weight ho

weights x inputs garda -> 3 x 4   le 4 x 1 Sanga np.dot() garda 3x1 ko Matrix Pauchu Where
Each Row is Each Neuron ko weighted sum ho

inputs 1 x 4 bhayeni np.dot(weights , inputs) garda -> inputs lai 4 x 1 banauara Matrix Multiply Garcha

"""

weighted_sum = np.dot(weights , inputs)
print(weighted_sum)

output = np.dot(weights , inputs ) + biases #Yo Direct Garda np.dot() bata aako result le same index ko sanga
#add garera result dincha
print(output)
```

<h3 style="background-color:purple; color:white; padding:10px; border-radius:8px; text-align:center;">
     Layer Of Neurons & Batch Of Data using Numpy
</h3>


```python
inputs = np.array([
    [1,2,3,2.5],
    [2,4,-1,2.0],
    [-1.5,2.7,3.3,-0.8]
])

weights = np.array([
    [0.2,0.8,-0.5,1],
    [0.5,-0.91,0.26,-0.5],
    [-0.26,-0.27,0.17,0.87]
])

biases = np.array([ 1 , 2 , 3])

output = np.dot(inputs , weights.T) + biases #.T le Transpose garcha Tara Array lai matra List lai haina
print(output)
```
