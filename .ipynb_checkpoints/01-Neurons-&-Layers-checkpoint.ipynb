<h1 style="background-color:purple; color:white; padding:10px; border-radius:8px; text-align:center;">
  Neuron
</h1>


```python
#Inputs , Bias ko List Banaune

#Neuron with 3 inputs
inputs = [1,2,3]
#Here x1 is 1 , x2 is 2 & x3 is 3
weights = [0.6 , 0.7 , 0.2 ] #Here w1 is 0.6 , w2 is 0.7 & w3 is 0.2
bias = 2

output = (inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + bias)
#List ko Indexing starts from 0 so x1 = inputs[0]
print(output)
```

```python
inputs = [ 4 , 5 , 2  , 1]
weights = [0.1 , 0.3 , 0.5 , 0.6 ]
bias = 2

output = inputs[0] * weights[0] + inputs[1] * weights[1] + inputs[2] * weights[2] + inputs[3] * weights[3] + bias
print(f"Output of Neuron with 4 inputs is {output}")
```

<h1 style="background-color:red; color:white; padding:10px; border-radius:8px; text-align:center;">
  Layer of Neurons
</h1>


```python
#Layer ma Multiple Neurons Huncha

inputs = [ 4 , 5 , 6  , 7]
weights = [
    #Every Neuron Sanga ko Afnai Associated Weights cha
    [ 0.3 , 0.4 , 0.9 , 0.1 ] , #Neuron1 sanga associated weights
    [ 0.4 , 0.5 , 0.1, 0.8 ],
    [0.5, 0.7 , 0.8, 0.4  ],
]

weight1 = weights[0] #LIST OF WEIGHTS ASSOCIATED WITH 1st NEURONS : W11 , W12 , W13 , W14
weight2 = weights[1] #W21 , W22 , W23 , W24
weight3 = weights[2] #W31, W32 , W33 , W34

biases = [ 2, 6 , 1]
b = biases

outputs = [
    #Neuron1 ko Output
    weight1[0] * inputs[0] + weight1[1] * inputs[1] + weight1[2] * inputs[2]  + weight1[3] * inputs[3] + b[0],

    #Neuron2 ko Output
    weight2[0] * inputs[0] + weight2[1] * inputs[1] + weight2[2] * inputs[2] + weight2[3] * inputs[3] + b[1],

    #Neuron3 ko Output
    weight3[0] * inputs[0] + weight3[1] * inputs[1] + weight3[2] * inputs[2] + weight3[3] * inputs[3] + b[2]
]

print(f" Neuron1 ko Output : {outputs[0]}\n Neuron2 ko Output : {outputs[1]}\n Neuron3 ko Output : {outputs[2]}")
    
```

<h2 style="background-color:#4CAF50; color:white; padding:10px; border-radius:8px; text-align:center;">
  Using Loops For Better and Easier Coding
</h2>


```python
inputs = [1.1, 2.2, 3.3]

# 10 neurons, each with 3 weights
weights = [
    [2, 0, 0],
    [2, 2, 0],
    [2, 2, 2],
    [1, 1, 1],
    [0.5, 0.5, 0.5],
    [1, 2, 3],
    [0, 1, 2],
    [3, 2, 1],
    [0.1, 0.2, 0.3],
    [5, 4, 3]
]

biases = [0, 0, 1, 1, 0.5, 2, 1, 0, 0.1, 10]

outputs_of_layer = []


#-> Since Weight ko 1st Subscript le Neuron ko Index bhancha so 1st Loop biases ko Length Samma gareko
#-> Since jati ota bias teti ota Neuron
for i in range(0 , len(biases)):
    output = 0

    #-> 2nd Subscript le Input ko Index bhancha so Loop inputs ko length samma
    for j in range(0 , len(inputs)):
        output += (weights[i][j] * inputs[j])
    output += biases[i]
    outputs_of_layer.append(output)

print("10 Neurons को Outputs:")
for index , output in enumerate(outputs_of_layer):
    print(f"Neuron {index + 1} को Output: {output}")
    
```
